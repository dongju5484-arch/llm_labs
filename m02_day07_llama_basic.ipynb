```python
from llama_index.llms.openai import OpenAI #OpenAI LLM모델을 불러온다
```

```python
import os
from dotenv import load_dotenv
```

```python
# 문서 로드 및 인덱스
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
```

```python
# env파일을 읽어서 환경변수에 api key 설정
load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
```

```python
# OpenAI GPT-4o 모델을 사용하도록 설정
llm = OpenAI(model='gpt-4o')
```

```python
# pdf 문서 폴더 불러오기
documents = SimpleDirectoryReader(input_dir='pdf1').load_data()
```

```python
documents
```

```python
# 백터스토어 인덱스 생성
index = VectorStoreIndex.from_documents(documents, llm=llm)
```

```python
index
```

```python
# 쿼리 엔진 설정
# as_query_engine() --> 기존의 인덱스를 검색할 수 있도록 쿼리 엔진으로 변환

# 쿼리 엔진 --> 사용자가 입력한 질문을 이해하고, 데이터에 접근해서 필요한 정보를 추출
#               유사성을 계산하여 가장 적합한 답변을 찾는 역할을 수행한다.
query_engine = index.as_query_engine()
```

```python
query_engine
```

```python
# 질문 입력
response = query_engine.query('미국의 인공지능 정책과 주요 변화에 대해 알려주세요')
```
